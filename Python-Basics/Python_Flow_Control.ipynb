## Conditional Statements

Conditional statements in Python are used to execute certain blocks of code based on specific conditions. These statements help control the flow of a program, making it behave differently in different situations.

### Python if Statement
- In control statements, The `if` statement is the simplest form. It takes a condition and evaluates to either `True` or `False`.
- If the condition is True, then the True block of code will be executed, and if the condition is False, then the block of code is skipped, and The controller moves to the next line.

### Python if...else Statement
- The `if-else` statement checks the condition and executes the if block of code when the condition is True, and if the condition is False, it will execute the else block of code.
<img src="../images/if-else.png" alt="" width="400" height="100"/>

```python
# if condition:
    # body of if statement

# else:
     # body of else statement
```

```python
number = int(input('Enter a number: '))

if number > 0:
    print('Positive number')
else:
    print('Not a positive number')

print('This statement always executes')
```

```python
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("Given number is even")
else:
    print("Given number is odd")
```

### Chain multiple if statement in Python

- In Python, the `if-elif-else` condition statement has an `elif` blocks to chain multiple conditions one after another. This is useful when you need to check multiple conditions.
- With the help of `if-elif-else` we can make a tricky decision. The `elif` statement checks multiple conditions one by one and if the condition fulfills, then executes that code.

```python
# if condition-1:
#      statement 1
# elif condition-2:
#      stetement 2
# elif condition-3:
#      stetement 3
#      ...
# else:
#      statement
```

```python
choice == 1
```

```python
choice = int(input("Enter your choice: "))

if choice == 1:
    print("Admin")

elif choice == 2: # else if
    print("Editor")

elif choice == 3:
    print("Guest")

else:
    print("Wrong entry")

print("I always run")
```

### Python for Loop

- Using for loop, we can iterate any sequence or iterable variable.
- The sequence can be string, `list`, `dictionary`, `set`, or `tuple`.

```python
# for element in sequence:
#      body of for loop
```

```python
groceryItems = ['Milk', 'Eggs', 'Bread', 'Butter']


for element in groceryItems:
    print(element + " is in my grocery list")

```

```python
nums = [1,2,32,34]

for num in nums:
    print(num**3)
```

### For Loop with Python range()
#### In Python, the range() function returns a sequence of numbers

#### The `range()` function takes up to three integer arguments:
`range(start, stop, step)`
- start (optional): The starting number of the sequence (inclusive). The default is 0 if not specified.
- stop (required): The number before which the sequence must end (exclusive). This value is never included in the generated sequence.
- step (optional): The difference between each number in the sequence. The default is 1 if not specified. This value cannot be 0

```python
len(num_list)
```

```python
num_list = [1,2,3,4,4,6]

for num in range(len(num_list)): # range(0, 6)
    print(num_list[num])
```

```python
for numbers in range(0, 20, 2): # range(0, 10)
    print(numbers)
```

```python
list(range(2,20))
```

```python
list(range(10, 2, -2))
```

```python
for numbers in range(10,21):
    print(numbers)
```

### While Loop

- In Python, The while loop statement repeatedly executes a code block while a particular condition is true.

- In a while-loop, every time the condition is checked at the beginning of the loop, and if it is true, then the loop’s body gets executed. When the condition became False, the controller comes out of the block.

<img src="../images/Python-while-Loop-Syntax.png" alt="" width="800" height="300"/>

```python
# while condition :
#     body of while loop
```

```python
number = -1

while number <= 5:
    print(number)
    number = number + 1 # number+=1
```

```python
for number in range(1,6):
    print(number)
```

```python
num = 10
sum_val = 0
i = 1
while i <= num:
    sum_val = sum_val + i
    print(sum_val)
    i = i + 1
print("Sum of first 10 number is:", sum_val)
```

### break and continue Statement
#### The `break` and `continue` statements are used to alter the flow of loops.

#### The break Statement
- The `break` statement terminates the for loop immediately before it loops through all the items.

```python
languages = ['Swift', 'Python', 'Go', 'C++']

for lang in languages:
    if lang == 'Swift':
        # print(lang)
        break
    print(lang)
```

#### The continue Statement
- The `continue` statement skips the current iteration of the loop and continues with the next iteration.

```python
languages = ['Swift', 'Ruby','Python','Java' ,'Go', 'C++']

for lang in languages:
    if lang == 'Ruby':
        continue
    print(lang)
```

```python
languages = ['Swift', 'Ruby','Python','Java' ,'Go', 'C++']

for lang in languages:
    pass


# custom exceptions
```

### Nested for loops
1. A loop can also contain another loop inside it. These loops are called nested loops.
2. In a nested loop, the inner loop is executed once for each iteration of the outer loop.

```python
# outer loop
attributes = ['Electric', 'Fast']
cars = ['Tesla', 'Porsche', 'Mercedes']

used_attributes = attributes[0]

for car in cars:
    print(used_attributes, car)

# for attribute in attributes: # O(n)
#     for car in cars: # O(n)
#         print(attribute, car)
```

```python
for a,b in zip(attributes,cars):
    print(a,b)
```
