## 🐍 Python Strings


#### 📌 What is a String?


- A string in Python is a sequence of characters enclosed in single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

```python
name = "Python"

city = 'Toronto'

message = "this is a single line string"


message = '''This is a
second-line string
third-line string
'''
```

### 📌 Characteristics of Strings
- Strings are immutable (cannot be changed after creation)
- Indexed (each character has a position)
- Can contain letters, numbers, spaces, and symbols

```python
name = "sahil"
name[2] = 'j'

name = "sajil"
```

```python
text = "Hello"
print(text[4])
# text[0] = 'h' ❌ Not allowed
```

```python
name = "432y78352jb5j2njnjkn542nj2n4j@@@@---"
name = "432y78352jb5j2njnjkn542nj2n4j@@---"
```

### 📌 String Indexing
- Each character has an index starting from 0.

<img src="../images/python-strings.png" alt="" width="800" height="300"/>

```python
word = "Python"
print(word[0]) # P
print(word[3]) # h
print(word[-1]) # n
```

### 📌 Common String Functions (Methods)
#### 1️⃣ upper() :

Converts string to uppercase.

```python
word = "python"

word_capital = word.upper()
print(word)
print(word_capital)
```

#### 2️⃣ lower() :

Converts string to lowercase.

```python
text = "PYTHON"
print(text.lower())
```

#### 3️⃣ strip() :

Removes leading and trailing spaces.

```python
text = "  hello "
print(text)
print(text.strip())
```

#### 4️⃣ replace(old, new):

Replaces part of a string.

```python
text = "I like Java"

text.
new_text = text.replace("Java", "Python")

print(text)
```

#### 5️⃣ split()
Splits string into a list.

```python
text = "apple banana apple" # --> list

text.split(" ")
# print(text.split(" "))
```

#### 6️⃣ join()
Joins list elements into a string.

```python
words = ["I", "love", "Python"]
print(" ".join(words))


joined_words = " ".join(words)
joined_words
```

#### 7️⃣ find()
Finds index of substring.

```python
text = "Python Programming"
print(text.find("Prog"))
```

```python
sentence = "sandy is sharp than sandy"

sentence_list = sentence.split(" ")
# print(sentence_list)

for element in sentence_list:
    if element == "sandy":
        print(sentence_list.index(element))
        continue

```

#### 8️⃣ count()
Counts occurrences of a character or word.

```python
text = "banana"
print(text.count("z"))
```
