### Types of Python Function Arguments

##### Python supports various types of arguments that can be passed at the time of the function call. In Python, we have the following function argument types in Python:

1. Default argument
2. Keyword arguments (named arguments)
3. Positional arguments
4. Arbitrary arguments (variable-length arguments *args and **kwargs)

#### Default Arguments

* A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument. The following example illustrates Default arguments to write functions in Python.

````python
def myHoliday(movie,food="Pizza"):
    return "My holiday movie is ",movie,"and my favorite food is ",food
````

````python
myHoliday(movie="Harry Potter",food="Dosa")
````

#### Keyword Arguments

* The idea is to allow the caller to specify the argument name with values so that the caller does not need to remember the order of parameters.

````python
# TODO : myholiday with keyword param

def learning(name,course,language="Python"):
    return f"My name is {name} and I am learning {course} using {language}"
````

````python
learning(name="Alex",course="Data Science",language="R")
````

````python
names = ["Alex","Yogesh"]

for name in names:
    print(learning(name=name,course="Data Science",language="R"))
````

````python
learning(course="Data Science",name="Alex")
````

````python
learning(course="Data Engineering",name="Yogesh")
````

#### Arbitrary Keyword Arguments

* It refers to the ability of a function to accept an unspecified number of keyword arguments. This is achieved using the `*args` and `**kwargs` syntax within the function's parameter list.

#### Why do we need Arbitrary Keyword Arguments?

* Sometimes, you might not know in advance how many keyword arguments a function will receive.
* `*args` and `**kwargs` allows you to collect these arguments into a list or dictionary within the function.
* The `*args` variable becomes a list,string,tuple etc and the `*kwargs` variable becomes a dictionary where the keys are the argument names and the values are the argument values.

#### *args (Positional Arguments)

Use `*args` when you want to pass a variable number of non-keyworded arguments (a list of items). Inside the function, args becomes a tuple.

````python
nums = [1,2,3,4]

sum_val = 0

for elements in nums:
    print(elements)
    sum_val += elements # sum_val = sum_val + elements

print(sum_val)
````

````python
def addingElementsList(*args):

    for elements in args:
        print(len(elements))
````

````python
addingElementsList([1,2,3,4],[1,2,34,5])
````

````python
def sumList(myList):
    total = 0
    for numbers in myList:
        total += numbers
    return total


myList = [1, 2, 3, 4, 5,8,3,4,2,43]

sumList(myList=myList)
````

#### **kwargs (Keyword Arguments)

Use `**kwargs` when you want to pass a variable number of keyworded arguments (named variables like key="value"). Inside the function, kwargs becomes a dictionary.

````python
def student_info(**kwargs):
    print(kwargs)


student_info(name="John", age=20, grade="A", learning="Python", myKey=90,myfavstudent="Tarun")
````

````python
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")


student_info(name="Bob", age=19, subject="Math", myKey=80,myfavstudent="Tarun")
````

````python
def student_info(**kwargs):
    print("Name:", kwargs.get("name"))
    print("Age:", kwargs.get("age"))


student_info(name="Alice", age=22)
````

````python
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")
````

````python
sahil_info = {
    "name":"Sahil",
    "age":22
}
````

````python
student_info(**sahil_info)
````
