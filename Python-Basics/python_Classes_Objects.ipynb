### Python Classes
* A class is considered a blueprint of objects.

* We can think of the class as a sketch (prototype) of a house. It contains all the details about the floors, doors, windows, etc. Based on these descriptions, we build the house; the house is the object.

* Since many houses can be made from the same description, we can create many objects from a class.

```python
# define a class
class Dog:

    sound = "bark"  # class attribute

    def speak(self): # class method
        print("Dogs make sound : ", self.sound)
```

### Python Objects

* Suppose Bike is a class then we can create objects like bike1, bike2, etc from the class.

```python
# define a class
class Dog:

    sound = "bark"  # class attribute

    def speak(self): # class method
        return "Dogs make sound : ", self.sound

# Creating object from class
dog_1_object = Dog()
```

### init method-constructor

- When you create an object from a class, it’s like building a house from a blueprint.
- The __init__() method is like the setup/initialization step where you decide the color of the walls, number of rooms, etc.
- The __init__() method is called automatically when you create an object from a class.
- Without __init__(), every house would be blank — you’d have to set values one by one after construction.

```python
# Simple Class Without __init__()

class Student:

    name = "Sahil"
    course = "Python"

    def learning(self):
        return f"{self.name} is learning {self.course}"



sahil_obj = Student()
alice_obj = Student()

# You now have empty Student objects
alice_obj.learning()
```

1. __init__ with Parameters

- You can pass multiple parameters to set up different attributes.

```python
class Person:
    def __init__(self, name, course):
        self.name = name # self.name
        self.course = course

    def learning(self):
        return f"{self.name} is learning {self.course}"


sahil_obj = Person(name="sahil",course = "Python")

alice_obj = Person(name = "Alice", course = "Java")


alice_obj.learning()
```

Explanation:

- **self**: refers to the current object (always the first parameter).
- **name** and **age**: parameters passed when creating the object.
- **self.name** and **self.age**: object attributes that store these values.

2. Default Parameters in __init__

Like normal functions, you can use default values in constructors.

```python
class Dog:
    def __init__(self, name, breed="Mixed", age=1): # class method
        self.name = name # class attributes
        self.breed = breed # class attributes
        self.age = age# class attributes

    def speak(self,color="black"): # function attributes # # class method
        return f"{self.name} is a {self.age} year old {self.breed} and have {color} color"

d1 = Dog(name = "Buddy")
d2 = Dog(name = "Max", breed = "Golden Retriever", age = 5)

d1.speak(color="white")
# print(d1.name, d1.breed, d1.age)
# print(d2.name, d2.breed, d2.age)
```

```python
class Rectangle:

    def __init__(self,width = 5.0,height = 10.0):
        self.width = width
        self.height = height

    def calculate_perimeter(self):
        return 2 * (self.width + self.height)

rectangle_one = Rectangle(width = 7.5,height = 12.0)
rectangle_one.calculate_perimeter()

rectangle_two = Rectangle()
rectangle_two.calculate_perimeter()


```
