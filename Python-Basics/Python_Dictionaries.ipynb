### Dictionary

#### Python dictionary is a data structure that stores the value in key: value pairs. Values in a dictionary can be of any data type and can be duplicated, whereas keys can't be repeated and must be immutable.

1. Keys are case sensitive which means same name but different cases of Key will be treated distinctly.
2. Keys must be immutable which means keys can be strings, numbers or tuples but not lists.
3. Duplicate keys are not allowed and any duplicate key will overwrite the previous value.
4. Internally uses hashing. Hence, operations like search, insert, delete can be performed in Constant Time.

```python
my_info = {
    'name' : 'Sahil',
    'age' : 28,
    'city' : 'Delhi',
    'country' : 'India',
    'new_age' : 29,
    (1,2):"sahil"
}
```

### Dictionary Methods

#### 📌 clear: Removes all key-value pairs from a dictionary, leaving it empty.


```python
my_dict = {
    'name': 'Alice',
    'age': 25,
    'city': 'New York',
    'food' : 'pizza',
    'phoneNumber' : 1234567890
}
print(my_dict)
my_dict.clear()
print(my_dict)  # Output: {}
```

#### 📌 copy : `dict.copy()` creates a **shallow copy** of a dictionary (a new dict with the same key→value references) so changes to the new dict don’t affect the original top-level keys.

```python
# copy
my_dict = {'name': 'Alice',
           'age': 25}


new_dict = my_dict.copy()




# # my_dict['age'] = 30
print(my_dict)

new_dict['age'] = 30
print(new_dict)
```

#### 📌 from keys : The fromkeys() method returns a dictionary with the specified keys and the specified value.

```python
# from keys
myKeys = ['x', 'y', 'z']
new_dict = dict.fromkeys(myKeys, 2)
print(new_dict)
```

```python
# walmart

product = ['cosmetics','dairy','bakery']
dict_product = dict.fromkeys(product,0)
dict_product

```

#### 📌 get : Using the .get() method: This method safely retrieves a value. It takes two optional arguments: the key to search for and an optional default value to return if the key is not found.

```python
my_dict = {'name': 'Alice', 'age': 25}
# print(my_dict['phone'])

print(my_dict.get('age','This key is not found'))
```

```python
# get
my_dict = {'name': 'Alice', 'age': 25}
print(my_dict.get('name'))
print(my_dict.get('salary', 0))
```

#### 📌 items : the items() method is a built-in dictionary function that returns a view object containing all the dictionary's key-value pairs as a list of tuples.

```python
# items

my_dict = {'name': 'Alice', 'age': 25}

# my_dict_count = Counter(numbers)
my_dict.items()

# [('name', 'Alice'), ('age', 25)]


```

#### 📌 keys : Returns a view object that displays a list of all the keys in the dictionary in order of insertion

```python
# keys
my_dict = {'name': 'Alice', 'age': 25,'phone':'13232'}
my_dict.keys()
```

#### 📌 values : Returns a view object containing all dictionary values, which can be accessed and iterated through efficiently


```python
# values
my_dict = {'name': 'Alice', 'age': 25,'phone':''}
my_dict.values()
```

#### 📌 pop : Returns and removes the element with the given key

```python
# pop
my_dict = {'name': 'Alice', 'age': 25}
value_deleted = my_dict.pop('age')

print(my_dict)
print(value_deleted)
```

#### 📌 popitem : Returns and removes the item that was last inserted into the dictionary.

```python
# pop item
my_dict = {'name': 'Alice', 'age': 25,'city':'New York'}
poppedItem = my_dict.popitem() # 1. delete the last item 2. return the deleted item
print(poppedItem)
print(my_dict)
```

#### setdefault() : Returns the value of a key if the key is in the dictionary else inserts the key with a value to the dictionary

when key exists

```python
d = {'a': 97, 'b': 98}

d
print("setdefault() returned:", d.setdefault('c', 99))
print("After using setdefault():", d)
```

when key not exists

```python
d = {'A': 'Geeks', 'B': 'For'}

print("Before setdefault():", d)

val = d.setdefault('C', 'Geeks')

print("Returned value:", val)
print("After setdefault():", d)
```
